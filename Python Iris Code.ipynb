```python
#dealing with outliers
def percentile(x) :
    Q1, Q3 = np.percentile(x, [25, 75])
    IQR = Q3 - Q1
    Upper = Q3 + 1.5*IQR
    Lower = Q1 - 1.5*IQR
    return Upper,Lower
```

```python
Upper_bound, Lower_bound = percentile(df["SepalWidthCm"])
```

```python
Upper_bound
```

```python
Lower_bound
```

```python
mask1 = df["SepalWidthCm"] <= Upper_bound
mask2 = df["SepalWidthCm"] >= Lower_bound
mask = mask1 & mask2
df = df[mask]
```

```python
sns.boxplot(y = "SepalWidthCm", data = df)
```

```python
import tkinter as tk
import numpy as np

# Function to handle prediction and display result in the Tkinter window
def predict_species():
    try:
        # Get input values from the user
        sepal_length = float(entry_sepal_length.get())
        sepal_width = float(entry_sepal_width.get())
        petal_length = float(entry_petal_length.get())
        petal_width = float(entry_petal_width.get())

        # Create the input array and perform scaling
        x_new = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
        x_new_scaled = scalar.transform(x_new)

        # Predict the class of the iris flower
        prediction = model_final.predict(x_new_scaled)

        # Display the result in the result label
        result_label.config(text=f"Predicted Species: {prediction[0]}")

    except ValueError:
        result_label.config(text="Invalid Input! Please enter valid numbers.")

# Initialize the main window
root = tk.Tk()
root.title("Iris Flower Classification")
root.geometry("600x400")
root.configure(bg="#f0f0f0")

# Title label
label_title = tk.Label(root, text="Iris Flower Classification", font=("Arial", 16, "bold"), bg="#f0f0f0", fg="#333")
label_title.pack(pady=10)

# Input fields
frame_inputs = tk.Frame(root, bg="#f0f0f0")
frame_inputs.pack(pady=10)

labels = ["Sepal Length (cm):", "Sepal Width (cm):", "Petal Length (cm):", "Petal Width (cm):"]
entries = []

# Dynamically create labels and entry fields
for i, text in enumerate(labels):
    label = tk.Label(frame_inputs, text=text, font=("Arial", 12), bg="#f0f0f0")
    label.grid(row=i, column=0, padx=10, pady=5, sticky="w")

    entry = tk.Entry(frame_inputs, font=("Arial", 12))
    entry.grid(row=i, column=1, padx=10, pady=5)
    entries.append(entry)

# Assign entries to variables for readability
entry_sepal_length, entry_sepal_width, entry_petal_length, entry_petal_width = entries

# Predict button
btn_predict = tk.Button(root, text="Predict Species", font=("Arial", 14), bg="#4CAF50", fg="white", command=predict_species)
btn_predict.pack(pady=20)

# Label to display the prediction result
result_label = tk.Label(root, text="", font=("Arial", 14), bg="#f0f0f0", fg="blue")
result_label.pack(pady=10)

# Start the Tkinter event loop
root.mainloop()
```
